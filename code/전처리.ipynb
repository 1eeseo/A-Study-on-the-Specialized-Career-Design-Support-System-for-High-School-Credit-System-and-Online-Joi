```python
address = pd.DataFrame()
address['주소'] = df['주소'].str.replace(r'\([^()]*\)\s*$', '', regex=True)
address
```

```python
import re


def clean_address(addr):
    # 1. 숫자-숫자 또는 숫자 다음 한글 제거 (ex: 183-4지평선고등학교 → 183-4)
    addr = re.sub(r'(\d+-?\d*)([가-힣]+)', r'\1', addr)

    # 2. 마침표(.) 이후 텍스트 제거 (ex: 551-23. 고산고등학교 → 551-23)
    addr = re.sub(r'\.\s*[^ ]*', '', addr)

    # 3. "고등학교", "중학교" 등 명칭 제거
    addr = re.sub(r'(고등학교|중학교|초등학교)', '', addr)

    # 4. 불필요한 이중 공백 제거
    addr = re.sub(r'\s+', ' ', addr).strip()

    return addr

address['주소'] = address['주소'].apply(clean_address)
```

```python
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut
import pandas as pd
import time

# 지오코더 설정
geo_local = Nominatim(user_agent='South Korea', timeout=10)

# 위도, 경도 반환 함수 정의
def geocoding(address):
    try:
        geo = geo_local.geocode(address)
        if geo:
            return [geo.latitude, geo.longitude]
        else:
            return [None, None]
    except GeocoderTimedOut:
        time.sleep(1)
        return geocoding(address)  # 재시도
    except:
        return [None, None]

# 주소 리스트 (예: df['클린주소'])
address_list = address['주소']

# 위도/경도 리스트 초기화
latitude = []
longitude = []

# 주소 하나씩 변환
for i in address_list:
    lat, lon = geocoding(i)
    latitude.append(lat)
    longitude.append(lon)

# 데이터프레임에 추가
address['위도'] = latitude
address['경도'] = longitude
```

```python
address_df = pd.DataFrame({'학교명' : df['학교명'], '주소' : address['주소'], '위도' : latitude, '경도':longitude})
address_df
```

```python
# 지역 목록에서 고유 색상 매핑
지역_목록 = new_df['행정구'].unique()
색상목록 = ['red', 'blue', 'green', 'purple', 'orange', 'darkred', 'lightblue',
           'lightgreen', 'gray', 'black', 'pink', 'cadetblue', 'darkblue']

# 지역명 → 색상 딕셔너리
지역색상 = {지역: 색상목록[i % len(색상목록)] for i, 지역 in enumerate(지역_목록)}
```

```python
import geopandas as gpd
import folium

# 1. SHP 파일 불러오기 (shp, shx, dbf 같은 파일들이 같은 폴더에 있어야 함)
gdf = gpd.read_file('/content/drive/MyDrive/교육공공 데이터 분석 공모전/data/고교학점제/LARD_ADM_SECT_SGG_전북특별자치도/LARD_ADM_SECT_SGG_52_202504.shp', encoding='euc-kr')
```

```python
gdf
```

```python
import folium

m = folium.Map(location=[35.8, 127.1], zoom_start=9)

# 전북 경계 GeoJson 시각화
folium.GeoJson(
    gdf,
    name="전라북도 경계",
    style_function=lambda feature: {
        'fillColor': 'none',
        'color': 'black',
        'weight': 1.5,
        'opacity': 0.8
    }
).add_to(m)
```

```python
# 학교별 마커 추가
for i, row in new_df.iterrows():
    folium.Marker(
        location=[row['위도'], row['경도']],
        popup=row['학교명'],
        tooltip=row['학교명'],
        icon=folium.Icon(color=지역색상.get(row['행정구'], 'gray'))
    ).add_to(m)

# 저장 및 표시
m.save('전북_지역별_고등학교_지도.html')
m
```
